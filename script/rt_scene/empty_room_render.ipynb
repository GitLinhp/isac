```python
from sionna.rt import load_scene, ITURadioMaterial, SceneObject
from isac.channel.rt import RT_SCENES_DIR
```

```python
scene = load_scene(
    str(RT_SCENES_DIR / "empty_room" / "empty_room.xml")
)
```

```python
cube_material = ITURadioMaterial(
    "cube-material", "metal", thickness=0.01, color=(0.8, 0.1, 0.1)
)

cube = SceneObject(fname=str(RT_SCENES_DIR / "cube.ply"), # Simple mesh of a car
                   name=f"cube",
                   radio_material=cube_material)

scene.edit(add=cube)
```

```python
scene.preview(clip_at=3.4)
```
