```python
scene = load_scene(
    str(RT_SCENES_DIR / "empty_room" / "empty_room.xml")
)
```

```python
sphere_material = ITURadioMaterial(
    "sphere-material", "metal", thickness=0.01, color=(0.8, 0.1, 0.1)
)

sphere = SceneObject(fname=sionna.rt.scene.sphere, # Simple mesh of a car
                     name=f"sphere",
                     radio_material=sphere_material)

scene.edit(add=sphere)
sphere.position = [0, 0, 1]
```
